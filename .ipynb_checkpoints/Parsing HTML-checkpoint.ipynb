```python
import urllib.request
with urllib.request.urlopen('https://www.rbc.ru/search/?query=%D0%93%D0%B0%D0%B7%D0%BF%D1%80%D0%BE%D0%BC') as f:
    html = f.read().decode('utf-8')
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html)
```

```python
for tag in soup.find_all('a'):
    print(tag)
```

```python
with urllib.request.urlopen('https://api.vk.com/method/wall.get?access_token=247f3d70247f3d70247f3d70d6276dfd5c2247f247f3d70478c4cc7443965779421d6d7&owner_id=-25232578&v=5.131') as f:
    json_string = f.read().decode('utf-8')
```

```python
import json
json_res = json.loads(json_string)['response']
for item in json_res['items']:
    for attachment in item['attachments']:
        link = attachment.get('link')
        if link:
            print(link['url'])
            with urllib.request.urlopen(link['url']) as f:
                html = f.read().decode('utf-8')
            soup = BeautifulSoup(html)
            title_html = soup.find("div", class_="article__header__title")
            if title_html:
                title = title_html.h1.string
                print(str(title).strip())
```
